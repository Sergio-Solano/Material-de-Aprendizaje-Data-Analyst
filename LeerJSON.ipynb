```python
df_transcripcion = []

def traer_respuesta(texto_buscado, transcripcion_df, fila, nombre_columna):
    # Busca la respuesta a partir del texto específico en el mensaje
    indices = transcripcion_df[
        transcripcion_df['message'].str.contains(texto_buscado, na=False, regex=False)
    ].index.tolist()

    if indices and indices[-1] + 1 < len(transcripcion_df):
        fila[nombre_columna] = transcripcion_df.loc[indices[-1] + 1, 'message']
    else:
        fila[nombre_columna] = np.nan

    return fila

def traer_intencion_y_respuesta(texto_intencion, texto_respuesta, transcripcion_df, fila):
    # Buscar la pregunta de intención
    indices_intencion = transcripcion_df[
        transcripcion_df['message'].str.contains(texto_intencion, na=False, regex=False)
    ].index.tolist()

    # Si no hay intenciones, devolvemos una fila con NaN en las columnas de intención y respuesta
    if not indices_intencion:
        fila['Intencion'] = np.nan
        fila['Respuesta_Inquietud'] = np.nan
        return [fila]

    filas = []

    for idx_intencion in indices_intencion:
        # El siguiente mensaje es la intención
        if idx_intencion + 1 < len(transcripcion_df):
            nueva_fila = fila.copy()
            nueva_fila['Intencion'] = transcripcion_df.loc[idx_intencion + 1, 'message']
            
            # Buscar la respuesta posterior "¿Resolví tu inquietud?"
            respuesta = np.nan  # Si no encontramos la respuesta, se deja NaN
            for idx_respuesta in range(idx_intencion + 1, len(transcripcion_df)):
                if texto_respuesta in transcripcion_df.loc[idx_respuesta, 'message']:
                    respuesta = transcripcion_df.loc[idx_respuesta + 1, 'message'] if idx_respuesta + 1 < len(transcripcion_df) else np.nan
                    break
            
            nueva_fila['Respuesta_Inquietud'] = respuesta
            filas.append(nueva_fila)

    return filas

# --- PROCESO ---
for archivo in os.listdir(carpeta):
    if archivo.endswith('.json'):
        with open(os.path.join(carpeta, archivo), 'r', encoding='utf-8') as f:
            data = json.load(f)

        transcripcion = data.get('conversacion', {}).get('transcripcion', [])
        transcripcion_df = pd.DataFrame([
            {
                'contactId': data.get('contactId'),
                'sender': linea.get('sender'),
                'message': linea.get('message'),
                'timestamp': linea.get('timestamp')
            }
            for linea in transcripcion
        ])

        # Fila base para este archivo
        fila = {'contactId': data.get('contactId')}
        fila = traer_respuesta('Selecciona tu tipo de documento:', transcripcion_df, fila, 'documento')
        fila = traer_respuesta('Digita tu número de cédula (solo números):', transcripcion_df, fila, 'Num_ID')

        # Obtener las intenciones
        # Agregar la respuesta a la inquietud para cada intención encontrada
        fila = {'contactId': data.get('contactId')}
        fila = traer_intencion_y_respuesta(
            'Escribe el número o el texto de la opción que quieres consultar:',
            '¿Resolví tu inquietud? 1- Sí 2- No',
            transcripcion_df,
            fila
        )

        df_transcripcion.extend(fila)

# Mostrar el DataFrame final
df_final = pd.DataFrame(df_transcripcion)
df_final
```
